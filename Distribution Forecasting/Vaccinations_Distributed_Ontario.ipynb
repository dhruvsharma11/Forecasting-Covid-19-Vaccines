## Vaccination Data

```python
data = pd.read_csv("Vaccinations_Distributed_Ontario.csv")


data = data.rename(columns={'Total Population': 'Total Distributed', 'Year': 'Date'})

data['Date'] = pd.to_datetime(data['Date'])

end_date = '2021-09-01'
data = data[data['Date'] <= end_date]

data.head()

data.tail()
```

```python
# # Plotting the data
plt.figure(figsize=(10, 6))

plt.plot(data['Date'], data['Change Distributed'], marker='o', linestyle='-')
plt.title('Change in Distributed Over Time')
plt.xlabel('Date')
plt.ylabel('Change in Distribution')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Holt's Method (Double Exponential Smoothing) - Days DataFrame 

```python
# importing for Exponential Smoothing 

from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python

# Set the Year column as the index
data.set_index('Date', inplace=True)

# Fit the model with annual seasonality
model = ExponentialSmoothing(
    data['Change Distributed'],
    trend='add',
    seasonal='add',
    seasonal_periods=4,
    damped = True
).fit()

# Forecast the next 16 quarters
forecast = model.forecast(15)
forecast
```

## Forecasted Dataframe

```python
start_date = pd.to_datetime('2021-09-01')
quarters = pd.date_range(start=start_date, periods=len(forecast), freq='Q')

# Create a DataFrame with the forecasted values and the corresponding dates
forecast_df = pd.DataFrame({
    'Date': quarters,
    'Total Distributed': forecast.values
})

forecast_df
```

## Value between 2024 March to 2025 March

```python
forecasted_value = forecast_df.loc[14, 'Total Distributed'] - forecast_df.loc[10, 'Total Distributed']

print("Yearly Distribution 2024-2025", forecasted_value)
```

```python
from matplotlib.ticker import FuncFormatter

def millions_formatter(x, pos):
    return f'{x / 1e6:.2f}K'

plt.figure(figsize=(10, 6))
plt.plot(forecast_df['Date'], forecast_df['Total Distributed'], marker='o', linestyle='-', markersize=5)
plt.title('Forecasted Change Distributed Over Time')
plt.xlabel('Date')
plt.ylabel('Change in Distributed')
plt.gca().yaxis.set_major_formatter(FuncFormatter(millions_formatter))
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
